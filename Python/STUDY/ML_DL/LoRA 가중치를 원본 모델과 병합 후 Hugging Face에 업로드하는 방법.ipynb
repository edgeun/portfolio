```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch

# Hugging FaceÏóêÏÑú ÏõêÎ≥∏ Gemma Î™®Îç∏ Îã§Ïö¥Î°úÎìú
base_model_name = "google/gemma-2-9b-it"  # ÏõêÎ≥∏ Gemma Î™®Îç∏ (ÏòàÏãú)
fine_tuned_lora = "edgeun/model_output"  # LoRA Í∞ÄÏ§ëÏπòÍ∞Ä ÏóÖÎ°úÎìúÎêú Hugging Face repo

# ÏõêÎ≥∏ Î™®Îç∏ Î°úÎìú
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    torch_dtype=torch.float16,  # FP16ÏúºÎ°ú Î°úÎìúÌïòÏó¨ Î©îÎ™®Î¶¨ Ï†àÏïΩ
    device_map="auto"
)

# LoRA Í∞ÄÏ§ëÏπò Ï†ÅÏö©
lora_model = PeftModel.from_pretrained(base_model, fine_tuned_lora)

# LoRA Í∞ÄÏ§ëÏπòÎ•º ÏõêÎ≥∏ Î™®Îç∏Í≥º Î≥ëÌï©
merged_model = lora_model.merge_and_unload()

# ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†Ä Î°úÎìú
tokenizer = AutoTokenizer.from_pretrained(fine_tuned_lora)

# Î≥ëÌï©Îêú Î™®Îç∏ Ï†ÄÏû•
merged_model.save_pretrained("gemma-2-9b-it-ai-counselor")
tokenizer.save_pretrained("gemma-2-9b-it-ai-counselor")

print("‚úÖ LoRA Î≥ëÌï© ÏôÑÎ£å: gemma-2-9b-it-ai-counselor Ìè¥ÎçîÏóê Ï†ÄÏû•Îê®")
```

```python
repo_id = "edgeun/gemma-2-9b-it-ai-counselor"  # ÏóÖÎ°úÎìúÌï† Î™®Îç∏ repo ID

# Î≥ëÌï©Îêú Î™®Îç∏ÏùÑ Hugging FaceÏóê ÏóÖÎ°úÎìú
merged_model.push_to_hub(repo_id)
tokenizer.push_to_hub(repo_id)

print(f"üöÄ Î≥ëÌï©Îêú Î™®Îç∏Ïù¥ Hugging FaceÏóê ÏóÖÎ°úÎìúÎêòÏóàÏäµÎãàÎã§: https://huggingface.co/{repo_id}")
```
