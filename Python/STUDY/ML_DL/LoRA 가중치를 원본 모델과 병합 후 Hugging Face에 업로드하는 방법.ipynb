```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch

# Hugging Face에서 원본 Gemma 모델 다운로드
base_model_name = "google/gemma-2-9b-it"  # 원본 Gemma 모델 (예시)
fine_tuned_lora = "edgeun/model_output"  # LoRA 가중치가 업로드된 Hugging Face repo

# 원본 모델 로드
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    torch_dtype=torch.float16,  # FP16으로 로드하여 메모리 절약
    device_map="auto"
)

# LoRA 가중치 적용
lora_model = PeftModel.from_pretrained(base_model, fine_tuned_lora)

# LoRA 가중치를 원본 모델과 병합
merged_model = lora_model.merge_and_unload()

# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(fine_tuned_lora)

# 병합된 모델 저장
merged_model.save_pretrained("gemma-2-9b-it-ai-counselor")
tokenizer.save_pretrained("gemma-2-9b-it-ai-counselor")

print("✅ LoRA 병합 완료: gemma-2-9b-it-ai-counselor 폴더에 저장됨")
```

```python
repo_id = "edgeun/gemma-2-9b-it-ai-counselor"  # 업로드할 모델 repo ID

# 병합된 모델을 Hugging Face에 업로드
merged_model.push_to_hub(repo_id)
tokenizer.push_to_hub(repo_id)

print(f"🚀 병합된 모델이 Hugging Face에 업로드되었습니다: https://huggingface.co/{repo_id}")
```
